## 1. Import Required Libraries

The following libraries are required to evaluate the trained machine learning model.

These libraries provide different evaluation metrics and visualization tools.

```python
import joblib

import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    classification_report,
    confusion_matrix,
    roc_curve
)

import warnings
warnings.filterwarnings("ignore")
```

### Interpretation

The required libraries have been successfully imported.

We are now ready to evaluate the trained model.

## 2. Load the Best Model

The best-performing model selected during the training phase is loaded from the saved file.

This model will now be evaluated on the testing dataset.

```python
model = joblib.load("../models/best_model.pkl")
```

### Interpretation

The trained model has been loaded successfully and is ready for evaluation.

## 3. Generate Predictions

The trained model is used to predict the loan status of the testing dataset.

Both class predictions and prediction probabilities are generated.

```python
y_pred = model.predict(X_test)

y_prob = model.predict_proba(X_test)[:,1]
```

### Interpretation

The model has generated predictions for all testing samples.

These predictions will now be evaluated using multiple performance metrics.

## 4. Accuracy Score

Accuracy measures the percentage of correctly classified observations.

Although accuracy is useful, it should not be the only metric for imbalanced datasets.

```python
accuracy = accuracy_score(y_test, y_pred)

print(f"Accuracy : {accuracy:.4f}")
```

### Interpretation

Accuracy indicates the overall correctness of the model.

However, credit risk datasets are often imbalanced, so additional metrics such as Recall and ROC-AUC are also important.

## 5. Precision Score

Precision measures how many applicants predicted as risky were actually risky.

Higher precision reduces false positive predictions.

```python
precision = precision_score(y_test, y_pred)

print(f"Precision : {precision:.4f}")
```

## 6. Recall Score

Recall measures how many actual risky applicants were correctly identified.

For credit risk prediction, Recall is one of the most important evaluation metrics.

```python
recall = recall_score(y_test, y_pred)

print(f"Recall : {recall:.4f}")
```

## 7. F1 Score

The F1 Score combines Precision and Recall into a single metric.

It provides a balanced measure of model performance.

```python
f1 = f1_score(y_test, y_pred)

print(f"F1 Score : {f1:.4f}")
```

## 8. ROC-AUC Score

The ROC-AUC Score measures the model's ability to distinguish between low-risk and high-risk applicants.

Higher values indicate better classification performance.

```python
roc = roc_auc_score(y_test, y_prob)

print(f"ROC AUC : {roc:.4f}")
```

## 9. Classification Report

The classification report summarizes Precision, Recall, F1-Score, and Support for each class.

It provides a detailed overview of model performance.

```python
print(classification_report(y_test, y_pred))
```

### Interpretation

The classification report provides detailed performance for each class.

It helps identify whether the model performs equally well for both loan repayment and loan default cases.

## 10. Confusion Matrix

A Confusion Matrix summarizes the prediction results.

It shows:

- True Positives
- True Negatives
- False Positives
- False Negatives

This helps identify the types of prediction errors made by the model.

```python
cm = confusion_matrix(y_test, y_pred)

plt.figure(figsize=(6,5))

sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues"
)

plt.xlabel("Predicted")

plt.ylabel("Actual")

plt.title("Confusion Matrix")

plt.show()
```

### Interpretation

The confusion matrix shows how many applicants were correctly and incorrectly classified.

For credit risk prediction, minimizing False Negatives is especially important because approving a risky applicant may lead to financial loss.

## 11. ROC Curve

The ROC Curve visualizes the trade-off between the True Positive Rate and False Positive Rate.

A curve closer to the top-left corner indicates better model performance.

```python
fpr, tpr, threshold = roc_curve(y_test, y_prob)

plt.figure(figsize=(7,5))

plt.plot(fpr, tpr, label=f"AUC = {roc:.3f}")

plt.plot([0,1],[0,1],"--")

plt.xlabel("False Positive Rate")

plt.ylabel("True Positive Rate")

plt.title("ROC Curve")

plt.legend()

plt.show()
```

### Interpretation

The ROC Curve demonstrates how well the model separates the two classes.

A larger area under the curve indicates stronger classification capability.
