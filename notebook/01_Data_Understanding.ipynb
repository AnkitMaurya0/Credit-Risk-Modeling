## 8. Check Missing Values

Missing values are empty entries in the dataset.

Machine Learning models cannot directly work with missing data.

Before training the model, we need to identify which columns contain missing values.

```python
missing = (
    df.isnull()
      .sum()
      .sort_values(ascending=False)
)

missing[missing > 0]
```

### Interpretation

Several features contain missing values.

In the preprocessing stage, we will decide whether to remove these columns or fill the missing values using appropriate techniques.

## 9. Check Duplicate Records

Duplicate records can reduce data quality and negatively affect model performance.

This step checks whether duplicate rows exist in the dataset.

```python
df.duplicated().sum()
```

### Interpretation

If duplicate records exist, they will be removed during preprocessing to improve data quality.

## 10. Explore the Target Variable

The target variable is the output that our machine learning model will predict.

For this project, the target variable is **loan_status**.

Understanding the distribution of different loan statuses helps us identify whether the dataset is balanced or imbalanced.

```python
df["loan_status"].value_counts()
```

### Interpretation

The dataset contains multiple loan status categories.

For binary classification, we will later select the appropriate categories (for example, **Fully Paid** and **Charged Off**) and convert them into numerical labels.

## 11. Percentage Distribution of Loan Status

Instead of viewing only the counts, we calculate percentages.

This provides a better understanding of class imbalance, which is important before model training.

```python
(
    df["loan_status"]
      .value_counts(normalize=True)
      .mul(100)
      .round(2)
)
```

### Interpretation

If one class contains significantly more records than the other, the dataset is imbalanced.

In the next stages of the project, we will use **SMOTE (Synthetic Minority Over-sampling Technique)** to balance the training data before building machine learning models.
